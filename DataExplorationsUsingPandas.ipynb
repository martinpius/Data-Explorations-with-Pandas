# Numeric data Exploration with Pandas

```python
# Import another dataset from the web URL which is related to drinks consuptions by countries
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head(5)
```

```python
drinks.columns
```

```python
#Re-index the data by countries and summarize the beer_servings
```

```python
drinks.set_index('country')
```

```python
drinks.beer_servings.groupby(drinks.continent).mean()
```

```python
#Summary stat for the drinks
```

```python
drinks.groupby(drinks.continent).mean()
```

```python
#Get the simple plot for the summary drinks
```

```python
drinks.groupby(drinks.continent).mean().plot(kind = 'hist')
```
